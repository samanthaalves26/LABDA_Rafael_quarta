# Web-Scraping

**Web scraping** ou raspagem de dados  um mtodo para extrair e armazenar dados de websites de forma **automatizada** e **estruturada**.
<div>

</div>

O processo de web scraping funciona da seguinte maneira:

- **Envio de Solicitao:** Um script ou programa envia uma requisio HTTP (como GET) ao servidor web para acessar uma pgina especfica.

- **Recebimento do HTML:** O servidor retorna o HTML da pgina solicitada, que contm todo o contedo da pgina, incluindo texto, imagens, links e outros elementos.

- **Parsing do HTML:** O script analisa o HTML recebido, utilizando bibliotecas como BeautifulSoup (em Python), para identificar e extrair os dados desejados, como textos dentro de tags especficas, tabelas, ou links.

- **Armazenamento:** Os dados extrados so organizados e armazenados em um formato estruturado, como CSV, JSON, ou diretamente em um banco de dados.

### Exemplo 2: Conhecendo o BeautifulSoup

#### Passo 1: instalar bibliotecas

```python
!pip install beautifulsoup4
```

#### Passo 2: Importar bibliotecas

```python
from bs4 import BeautifulSoup
```

#### Passo 3: Requisitar informaes de um website

```python
response = requests.get('https://www.cps.sp.gov.br')
```

#### Passo 4: Conectar o site com a biblioteca BeautifulSoup

```python
soup = BeautifulSoup(response.content ,'html.parser')
```

#### Passo 5: Visualizar HTML

```python
print(soup)
```

#### Passo 6: Visualizar elementos especficos

```python
print(soup.title)
```

```python
print(soup.find('title'))
```

```python
print(soup.find('p'))
```

#### Passo 7: Filtrar atributos

```python
barra_nav = soup.find('nav', attrs = {'class': 'main-menu-container menu-mobile-inativo'})
print(barra_nav)
```

```python
menu = barra_nav.find('ul', attrs = {'class': 'menu'})
#print(menu)
```

```python
pos = menu.find('li', attrs = {'id': 'menu-item-80637'})
#print(pos)
```

#### Passo 8: Listar links

```python
links = pos.find_all("a")
links
```

```python
len(links)
```

```python
print(links[2].text)
```

```python
print(links[2].attrs)
```

#### Passo 10: Visualizar atributos

#### Passo 11: Extrair atributos especficos

```python
print(links[2]['href'])
```

#### Passo 12: Extrair uma lista de atributos
Neste exemplo, iremos extrarir todos os links da barra de navegao

```python
for i in links:
    try:
        print(i['href'])
    except:
        print(i.text)
```

### Exemplo 3: Regras para encontrar elemento no BeautifulSoup
Neste exemplo, vamos aproveitar os passos do exerccio anterior
- Passo 1: Instalar bibliotecas
- Passo 2: Importar bibliotecas
- Passo 3: Requisitar informao de um website
- Passo 4: Conectar o site com a biblioteca BeautifulSoup

E vamos explorar outras regras para encontrar elementos (.find)
- por classe
- por id
- por atributos
- por textos
- por pedaos de texto

#### Passo 5: Encontrar elemento por atributo
*.find*(nome do atributo = dados do atributo)

```python
barra_nav2 = soup.find(class_ = 'main-menu-container menu-mobile-inativo')
print(barra_nav2)
```

```python
pos2 = soup.find(id = 'menu-item-80637')
print(pos2)
```

```python
atr = soup.find('img')
print(atr)
```

```python
# Encontrar por texto
texto_exato = soup.find(string = 'Mestrado Profissional em Gesto e Tecnologia em Sistemas Produtivos')
```

```python
import re

parte_do_texto = soup.find_all(string = re.compile('Mestrado'))
print(parte_do_texto)
```

#### Passo 6 Encontrar elemento que contm o texto da busca
Neste passo iremos usar as funes **parent** e **contents**

```python
parent_texto = texto_exato.parent
print(parent_texto)
```

```python
print(parent_texto.contents)
```
