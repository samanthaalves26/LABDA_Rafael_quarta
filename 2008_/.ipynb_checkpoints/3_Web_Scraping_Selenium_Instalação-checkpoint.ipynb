### Instruções importantes:

- Para fazer scraping com a biblioteca Selenium use, preferencialmente, o navegador Chrome.
- Com o Google Chrome instalado, acesse a ULR chrome://settings/help para ver a versão do seu navegador.

<div>
<img src=attachment:175a48eb-5402-40d8-a71f-fe9198308f13.png width="600"/>
</div>
O número sublinhado em vermelho é o dado mais importante sobre a versão do seu navegador.

- Para que o Selenium controle o seu navegador é preciso obter os parâmetros de navegação e para isso é necessário baixar webdriver do seu navegador.
- Se o número versão do seu navegador for 128, então você precisa baixar o driver, chamado de chromedriver, que comece com o prefixo correspondente. - 
Acesse https://chromedriver.chromium.org/downloads e baixe versão compatível com o seu navegador.
- Se não houver um chromedriver da versão exata da sua, pegue uma verão que tenha pelo menos o primeiro número igual.



Descompacte o arquivo .zip e cole na pasta do anaconda3.a:.


#### Passo 3: Abrir o Chromedriver

```python
driver = webdriver.Chrome() # Se uma nova janela do Chrome foi aberta, quer dizer que o selenium e o chromedriver estão funcionando corretamente.
```

#### Passo 4 (opicional): Atualizar webdriver automaticamente

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

# Informar o o diretório onde deve ser inserido o webdriver atualizado 
driver = webdriver.Chrome(service=Service(r'C:\seu_caminho\chromedriver.exe'))
```

#### Passo 5 (opicional): Alternativa para atualizar webdriver automaticamente

```python
# Remova o comentário(#) da linha abaixo caso esteja executando este script pela primeira vez
#!pip install webdriver-manager
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```
